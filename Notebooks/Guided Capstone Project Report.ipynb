# Enhanced Report: Recommendations for Big Mountain Resort

## Executive Summary


This report leverages a detailed pricing model to explore various scenarios for Big Mountain Resort, aiming to align ticket pricing with market standards and potentially enhance revenue. The analysis considers the resort's current facilities and compares them to competitors, identifying strategic opportunities for pricing and operational adjustments.

## Current Pricing and Market Position

Big Mountain Resort currently charges 81 dollars for an adult weekend ticket. Our model, which incorporates data from similar resorts and adjusts for market conditions, suggests that the facilities at Big Mountain could support a ticket price of approximately 95.87 dollars. This indicates potential underpricing and room for a price increase, provided that the market conditions and customer perceptions support such a change.

## Key Facility Comparisons

Big Mountain stands out in several facility features when compared to the market:

**Vertical Drop:** Competitively high but not the highest.


**Snow Making Area:** Among the top, enhancing snow reliability.



**Total Number of Chairs:** One of the highest, suggesting substantial lift capacity.


**Fast Quads:** Well-equipped, offering quick lift times which enhance customer experience.


**Runs:** Number of runs is competitive, offering diverse skiing opportunities.


**Longest Run and Skiable Terrain:** Impressive lengths and area, appealing to a wide range of skiers.


## Modeling Scenarios and Recommendations

### Scenario Analysis:


**Closing Runs:** Modeling suggests minimal impact on ticket pricing from closing up to 10 less popular runs, though the decision should also consider guest satisfaction and brand impact.

The model says closing one run makes no difference. Closing 2 and 3 successively reduces support for ticket price and so revenue. If Big Mountain closes down 3 runs, it seems they may as well close down 4 or 5 as there's no further loss in ticket price. Increasing the closures down to 6 or more leads to a large drop. 





**Enhancing Vertical Drop and Chairlifts:** Adding a run that increases the vertical drop by 150 feet with an additional chairlift could support a higher ticket price, improving both guest experience and revenue without requiring more snow coverage.

## Proposed Actions:

Increase Ticket Prices: Cautiously raise prices towards the modeled suggestion of $95.87, aligned with facility enhancements and market positioning.

Operational Adjustments: Consider implementing the vertical drop enhancement. It provides a good balance of investment to potential revenue increase.

Run Management: Conduct a pilot test by temporarily closing the least used runs, evaluating both operational savings and guest feedback.

## Financial Implications and Further Considerations

The increase in ticket prices should be communicated clearly to guests, highlighting enhancements and improvements in facilities. This approach helps justify the price increase, aligning it with guest expectations and market standards.

## Further Work:

Cost Analysis: More detailed cost data would help refine the pricing model and fully assess profitability.

Market Dynamics: Further research into guest demographics and preferences could tailor both marketing and operational strategies more effectively.

Tool Development: Develop a user-friendly tool for business analysts at the resort to use the model independently, allowing for regular adjustments and scenario testing without requiring deep technical expertise.


## Conclusion

Big Mountain Resort has the opportunity to adjust its pricing strategy to better reflect the value it offers. By carefully considering facility enhancements and operational changes, the resort can increase revenue while maintaining guest satisfaction. Future efforts should focus on refining cost data and integrating market research to further optimize pricing and operational strategies.
